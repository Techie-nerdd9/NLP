```python
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
text="Hi I'm Gaya3. How r u? How's ur day "
output1 = word_tokenize(text)
output2 = sent_tokenize(text)
print(output1)
print(output2)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
input = "This is the second session"
doc = nlp(input)
```

```python
for token in doc:
  print(token)
```

```python
for sent in doc.sents:
  print(sent)
```
